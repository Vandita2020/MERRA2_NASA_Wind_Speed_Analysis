```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../Final_csv/New_york.csv")
```

```python
data
```

```python
data.groupby(['lat','lon']).groups
```

```python
one_data = data[(data['lat']== -79.75964943) & (data['lon']== 42.1966383050001)]
```

```python
one_data
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(one_data['month'], one_data['wind_speed'], color ='maroon',width = 0.4)
plt.xlabel("Month")
plt.ylabel("Wind Speed")
plt.title("Monthly Wind Speed data for location " + str(-79.75964943) + ',' + str(42.1966383050001) + ' in New York')
plt.show()
```

```python
# maximum wind speed
max_wind_by_geom = data.groupby(['lon','lat']).mean().reset_index()
print(max_wind_by_geom['wind_speed'].max())
```

```python
# monthly mean wind speed 
monthly_avg = data.groupby('month').mean().reset_index()
print(monthly_avg)
plt.bar(monthly_avg['month'], monthly_avg['wind_speed'])
plt.title("Monthly average wind speed across New York")
plt.xlabel("Month")
plt.ylabel("Wind Speed")
plt.show()
```
