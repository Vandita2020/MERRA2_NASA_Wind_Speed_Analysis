# Combining data

```python
import pandas as pd
import numpy as np
```

```python
florida = pd.read_csv("../Final_csv/Florida.csv")
new_york = pd.read_csv("../Final_csv/New_york.csv")
north_carolina = pd.read_csv("../Final_csv/North_carolina.csv")
```

```python
final_file = pd.concat([florida, new_york, north_carolina])
```
