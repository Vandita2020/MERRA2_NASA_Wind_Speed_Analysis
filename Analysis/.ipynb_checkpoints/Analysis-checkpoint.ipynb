```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
from math import pi
from math import sqrt
from math import atan2
```

```python
from windrose import WindroseAxes
```

```python
df = pd.read_csv("../Final_csv/Combined_file.csv")
```

```python
df = df.dropna()
```

```python
df
```

```python
df['wind_dir_trig_to_degrees'] = df['wind_direction'] * 180/pi ## -111.6 degrees
df['wind_dir_trig_from_degrees'] = df['wind_dir_trig_to_degrees'] + 180
df['wind_dir_cardinal'] = 90 - df['wind_dir_trig_from_degrees']
```

```python
df
```

```python
ax = WindroseAxes.from_ax()
ax.bar(df['wind_dir_trig_from_degrees'], df['wind_speed'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
```
