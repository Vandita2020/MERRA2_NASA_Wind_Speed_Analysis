```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../Final_csv/Florida.csv")
```

```python
data
```

```python
one_data = data[(data['lat']== -87.454122211) & (data['lon']== 30.923300623)]
```

```python
one_data
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(one_data['month'], one_data['wind_speed'], color ='maroon',width = 0.4)
plt.xlabel("Month")
plt.ylabel("Wind Speed")
plt.title("Monthly Wind Speed data for location " + str(-87.454122211) + ',' + str(30.923300623) + ' in Florida')
plt.show()
```

```python
# maximum wind speed
max_wind_by_geom = data.groupby(['lon','lat']).mean().reset_index()
print(max_wind_by_geom['wind_speed'].max())
```

```python
# monthly mean wind speed 
monthly_avg = data.groupby('month').mean().reset_index()
print(monthly_avg)
plt.bar(monthly_avg['month'], monthly_avg['wind_speed'])
plt.title("Monthly average wind speed across Florida")
plt.xlabel("Month")
plt.ylabel("Wind Speed")
plt.show()
```
